```python
##Crear un DataFrame de un millón de registros
cantidad = 1000000
```

```python
import random
```

```python
#Dato Aleatorios
influencer = [
        "viajero_cancun", "playa.lovers", "cancun_lifestyle",
        "turismo_mex", "sunset.cancun", "hoteluxe_cancun",
        "foodie_cancun", "adventure_cancun","familia_en_cancun",
        "eco_turismo_cancun"
        ]
tipo_post = [
    "reel_turismo",
    "foto_playa",
    "carrusel_lifestyle",
]
```

```python
#Creacion del DataFrame
dfMillon = pd.DataFrame ({
    "influencer" : [random.choice(influencer) for _ in range(cantidad)],
    "likes" : [random.randint(1,10000) for _ in range(cantidad)],
    "comentarios" : [random.randint(1,1000) for _ in range(cantidad)],
    "fecha" : [random.randint(2020,2026) for _ in range(cantidad)],
    "tipo_post" : [random.choice(tipo_post) for _ in range(cantidad)]
})
```
