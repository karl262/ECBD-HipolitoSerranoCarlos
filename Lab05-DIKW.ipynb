```python
# Llenar los datos faltantes de la columna 'director' ya que tiene más de 2k registros vacíos
df["director"] = df["director"].fillna("Desconocido")
```

```python
# Llenar los datos faltantes para 'cast' ya que tiene mas de 800 vacios
df["cast"] = df["cast"].fillna("N/I")
```

```python
# Llenar datos faltantes para 'country' con mas de 800 vacios
df["country"] = df["country"].fillna("N/I")
```

```python
df.isnull().sum()
```

```python
# Se realiza un llenado de datos en lugar de un borrado de columna debido a que es información que es necesaria para datos de peliculas o series para tener como referencia.
# Se asignan valores 'Desconocido' o 'N/I' para que se muestren en las columnas
```

```python
# Grafica para visualizar cuantas series y peliculas hay
plt.figure(figsize=(10,5))
sns.countplot(df, x="type")
plt.title("Proporcion entre peliculas y series de television")
plt.xlabel("Tipo")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Grafica para visualizar los titulos por tipo
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)
plt.title("Cantidad de títulos por tipo")
plt.show()
```

```python
# Grafica para visualizar las clasificaciones
plt.figure(figsize=(10,5))
df['rating'].value_counts().plot(kind='bar')
plt.title("Distribución de clasificaciones")
plt.xlabel("Rating")
plt.ylabel("Cantidad")
plt.show()
```

```python
# La primera grafica permite visualizar la proporcion entre peliculas y series dentro del catalogo.
# Es útil porque muestra de manera inmediata cuál es el tipo de contenido predominante en el dataset.
```
